```python
# 词云展示
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from lxml import etree
from nltk.tokenize import word_tokenize

# 去掉停用词
def remove_stop_words(f): 
    stop_words = ['Market']
    for stop_word in stop_words:
        f = f.replace(stop_word, '')
    return f

def create_word_cloud(f):
    f = remove_stop_words(f)
    cut_text = word_tokenize(f)
    cut_text = " ".join(cut_text)
    wc = WordCloud(
        max_words=100,
        width=2000,
        height=1200,
    )
    wordcloud = wc.generate(cut_text)
    return wordcloud

def word_prepare(path):
    # 数据加载
    dataset = pd.read_csv(path,header=None,encoding = 'gbk')   #中文为避免乱码及报错，需设置编码格式encoding = 'gbk'
    #数据格式准备
    transaction = []
    for i in range(0,dataset.shape[0]):
        temp = []
        for j in range(0,20):
            item = str(dataset.iloc[i,j])
            if item != 'nan':
                temp.append(dataset.iloc[i,j])
        transaction.append(temp)
        
        all_word = ' '.join('%s' %item for item in transaction) #transaction = [[],[],[]...]
    return all_word

def main():
    path = 'H:\\SVW-Data analysis PPT\\Data_Engine_with_Python-master\\L4\\MarketBasket\\Market_Basket_Optimisation.csv'
    word_data = word_prepare(path)
    cloud_pig = create_word_cloud(word_data)
    # 绘制图片
    plt.imshow(cloud_pig)
    # 消除坐标轴
    plt.axis("off")
    # 展示图片
    plt.show()
    # 保存图片
    cloud_pig.to_file('H:\SVW-Data analysis PPT\Data_Engine_with_Python-master\L4\MarketBasket\MarketBasket_cloud_2.png')
    
main()
```
