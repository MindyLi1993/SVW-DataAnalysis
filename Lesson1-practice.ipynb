```python
# Action1：求2+4+6+8+...+100的求和，用Python该如何写
sum = 0
for i in range(2,101,2):
    sum = sum + i
print(sum)
```

```python
# Action2: 统计全班的成绩,描述 
import pandas as pd
import numpy as np

dict = {"Name":["张飞","关羽","刘备","典韦","许褚"],"语文":[68,95,98,90,80],"数学":[65,76,86,88,90],"英语":[30,98,88,77,90]}
df = pd.DataFrame(data=dict)

df.iloc[:,1:].agg([np.mean,np.max,np.min,np.std,np.var])

#总成绩排序，得出名次进行成绩输出
df.sum(axis=0)
```

```python
df["总分"] = df[["语文","数学","英语"]].sum(axis=1)
```

```python
df.sort_values("总分",ascending = False)
```
